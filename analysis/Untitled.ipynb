```python
dfs = {}
dfs["64"] = pd.read_csv('csv/run-Jan18_09-52-29_SSM16_s64_40ce024_base-tag-epoch_test_x_loss.csv')
dfs["256"] = pd.read_csv('csv/run-Jan21_08-45-24_SSM16_s256_ca76dd0_base-tag-epoch_test_x_loss.csv')
dfs["512"] = pd.read_csv('csv/run-Jan18_15-04-14_SSM16_s512_40ce024_base-tag-epoch_test_x_loss.csv')
dfs["1024"] = pd.read_csv('csv/run-Jan18_09-59-41_SSM16_s1024_40ce024_base_std7-tag-epoch_test_x_loss.csv')[:-1]
dfs["64 + 512"] = pd.read_csv('csv/run-Jan22_03-57-33_SSM16_s64-512_ac6311e-tag-epoch_test_x_loss.csv')
dfs["64 + 512 + 1024"] = pd.read_csv('csv/run-Jan27_08-37-25_SSM16_s64-512-1024_9adc149-tag-epoch_test_x_loss.csv')[:-30]

labels = ["64", "256", "512", "1024"]
plt.figure(figsize=(16, 6))
plt.title("learning curve", fontsize=16)
plt.ylim(113700, 114600)
plt.xlabel("epochs", fontsize=14)
plt.ylabel("Negative Log Likelihood", fontsize=14)
plt.grid(True)
for label in labels[::-1]:
    plt.plot(dfs[label]["Step"], dfs[label]["Value"], label=label)
plt.legend(fontsize=14)
plt.show()
```
